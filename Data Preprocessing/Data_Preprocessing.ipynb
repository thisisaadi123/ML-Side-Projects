# Data Preprocessing Tools

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
#pandas creates a dataframe - format of data
dataset=pd.read_csv('Data.csv')
#features-> the columns with which we are gonna predict something X
#dependent variable->the prediction / value / target which depends on the features y
#iloc stands for locate indices , take the indices we want and extract a subset of data from the dataset
X=dataset.iloc[:,:-1].values #we want all the rows , we want all the columns except the last column
y=dataset.iloc[:,-1:].values
```

```
print(X)
```

```
print(y)
```

## Taking care of missing data

```
#we replace the missing value with the average
#of all the values across the column
#we will use the simpleImputer class from the scikit learn libary to do it
#we import the class and create an instance of the class , the object will help us handle missing values
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X[:,1:3])
X[:,1:3]=imputer.transform(X[:,1:3])

#3 steps for imputer ->create imputer object ,fit method,update with transform method
```

```
print(X)
```

## Encoding categorical data

### Encoding the Independent Variable

```

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# 2. SETUP TRANSFORMER
# Encodes index 0 (Country), passes through the rest
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')

# 3. APPLY
X = np.array(ct.fit_transform(X))

# This will now correctly print 5 columns
print(X)
```

### Encoding the Dependent Variable

```
#we will encode the dependent variable purchased with another encoder
#label encoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
#y is not neccessarily needed to be a numpy array for future machine learning models
```

```
print(y)
```

## Splitting the dataset into the Training set and Test set

```
#train test split module from sklearn helps us achieving the 80-20 or 60-20-20 split
#feature scaling should be done after training and testing splits
#this is because test set is from a domain of values which arent known to the model
#we shouldnt have information leakage like doing feature scaling before train test split
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```
print(X_train)
```

```
print(X_test)
```

```
print(y_train)
```

```
print(y_test)
```

## Feature Scaling

```
#normalization is recommended whenever we have a normal distribution
#standardization works well all the time
#use standardization 95% of the time unless you have a really good normal distribution of data
#the standard scalar will only be fit for X_train as the data which is scaled shouldnt affect the test set in anyway
```

```
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

#we shouldnt apply scaling to dummy variables
#they are already in 0-1 range within the standard scaler range of -3 to 3
#if we apply standardization on this we will lose the interpretation like
#france is 001 germany is 010 spain is 100 etc we will lose this information in the process of standardization
#fit calculates the mean and standard deviation for the column and transform will apply the transformation to our dataset and make changes accordingly

X_train[:,3:]=sc.fit_transform(X_train[:,3:])
X_test[:,3:]=sc.transform(X_test[:,3:])
```
